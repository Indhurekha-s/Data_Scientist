```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
data = {
    "Math_Score": np.random.randint(40,100,30),
    "Science_Score":np.random.randint(35,95,30)
}
df= pd.DataFrame(data)
print("Dataset:\n",df)
```

```python
desc_stats = df.describe()
print("Descriptive Statistics:\n",desc_stats)
```

```python
central_tendency = {
    "Mean": df.mean(),
    "Median": df.median(),
    "Mode": df.mode().iloc[0]
}
print("\nCentral Tendancy:\n",pd.DataFrame(central_tendency))
```

```python
dispresion = {
    "Variance": df.var(),
    "Standard Deviation": df.std(),
    "Range": df.max() - df.min(),
    "IQR": df.quantile(0.75) - df.quantile(0.25)
}
print("\nDispresion Measures:\n",pd.DataFrame(dispresion))
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df["Science_Score"],bins=6, kde=True, color='red', edgecolor='black')
plt.title("Histogram of Science Scores")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Math_Score"],color='blue')
plt.title("Box Plot of Math Score")
plt.xlabel("Score")
plt.grid(True)
plt.tight_layout()
plt.show()
```
